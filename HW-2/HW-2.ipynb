To check whether the given combination of pressure and temperature are casuing the explosion, 
we are using the difference between temperature it is in certain tolerance we say that flame has started the explosion.


```python

import numpy as np
import cantera as ct
import matplotlib.pyplot as plt

gas = ct.Solution('gri30.yaml')
fuel_species = 'C3H8'
gas.set_equivalence_ratio(1.0, fuel_species, 'O2:1.0, N2:3.76')

T_array = np.arange(300, 2000,100)
p_array = np.arange(1*ct.one_atm, 100*ct.one_atm,5*ct.one_atm)

T_explosion = []
p_explosion = []
T_first_explosion = []
p_first_explosion = []

for T_value in T_array:
    first_exploded = False  # Flag to check if an explosion has occurred for this T_value

    for p_value in p_array:
        T_initial = T_value
        gas.TP = T_value, p_value
        r = ct.IdealGasConstPressureReactor(gas)
        sim = ct.ReactorNet([r])
        time = 0.0
        ntimes = 100000

        for n in range(ntimes):
            time += 1e-07
            sim.advance(time)

            # Check if an explosion occurs
            if np.abs(T_initial - gas.T) >10:
                # Store the first explosion point only
                if not first_exploded:  # Corrected variable name
                    T_first_explosion.append(T_value)  # Store the temperature for the first explosion
                    p_first_explosion.append(p_value)   # Store the pressure for the first explosion
                    first_exploded = True  # Mark that the first explosion has occurred

                T_explosion.append(T_value)
                p_explosion.append(p_value)
                break  

```

The induction time calculated as time required for the gradient of temperature takes place

```python
import sys
import numpy as np
import cantera as ct
import matplotlib.pyplot as plt


fixed_temperature = 1400  # K
pressures = [5, 10, 20, 50]  # atm

# Initialize list to store induction times
induction_times = []


plt.figure(figsize=(10, 6))

for p in pressures:
    gas = ct.Solution('gri30.yaml')
    fuel_species = 'C3H8'
    gas.set_equivalence_ratio(1.0, fuel_species, 'O2:1.0, N2:3.76')
    
    # Set temperature and pressure
    gas.TP = fixed_temperature, p * 101325  # Convert atm to Pa
    
  
    r = ct.IdealGasConstPressureReactor(gas)
    sim = ct.ReactorNet([r])
    
    
    time = 0.0
    dt = 1e-05
    ntimes = 1000
    
    states = ct.SolutionArray(gas, extra=['t'])
    
    gradT = np.zeros(ntimes)
    T_before = fixed_temperature
    
    # Induction time tracking
    induction_time = None

    # Advance the simulation
    for n in range(ntimes):
        time += dt
        sim.advance(time)
        states.append(r.thermo.state, t=time * 1e3)  # in ms

        # Calculate the temperature gradient
        gradT[n] = np.abs(T_before - gas.T) / dt
        T_before = gas.T

    # Identify the induction time as the time when the temperature gradient is maximum
    induction_time = states.t[np.argmax(gradT)]

    
    plt.plot(np.log(states.t), states.T, label=f'Pressure={p} atm')

    
    induction_times.append(induction_time)

# Add plot labels and title for temperature evolution
plt.xlabel('log(Time) (ms)')
plt.ylabel('Temperature (K)')
plt.title('Temperature Evolution Over Time at T=1400 K for Different Pressures')
plt.legend()
plt.grid()
plt.show()

# New figure for induction times
plt.figure(figsize=(10, 6))
plt.bar(pressures, induction_times, color='skyblue')
plt.xlabel('Pressure (atm)')
plt.ylabel('Induction Time (ms)')
plt.title('Induction Time at T=1400 K for Different Pressures')
plt.xticks(pressures)
plt.grid()
plt.show()
```

```python
import numpy as np
import cantera as ct
import matplotlib.pyplot as plt

# Set up the gas model
gas = ct.Solution('gri30.yaml')
fuel_species = 'C3H8'
gas.set_equivalence_ratio(1.0, fuel_species, 'O2:1.0, N2:3.76')

# Temperature and pressure ranges
T_array = np.arange(300, 2000, 100)  # K
p_array = np.arange(1 * ct.one_atm, 100 * ct.one_atm, ct.one_atm)  # Pa

# Lists to store explosion data
T_explosion = []
p_explosion = []
T_first_explosion = []
p_first_explosion = []

# Loop over temperatures
for T_value in T_array:
    first_exploded = False  # Flag to track the first explosion for each T_value

    # Loop over pressures
    for p_value in p_array:
        T_initial = T_value
        gas.TP = T_value, p_value
        r = ct.IdealGasConstPressureReactor(gas)
        sim = ct.ReactorNet([r])

        time = 0.0  # Reset time for each run
        exploded = False  # Flag to track explosion occurrence

        while time < 0.05:  # Simulate for a maximum time of 50 ms
            time = sim.step()  # Adaptive time stepping

            # Check for explosion (temperature rise > 10 K)
            if gas.T - T_initial > 10:
                if not first_exploded:
                    # Store the first explosion point
                    T_first_explosion.append(T_value)
                    p_first_explosion.append(p_value)
                    first_exploded = True  # First explosion recorded
                
                # Store all explosion points
                T_explosion.append(T_value)
                p_explosion.append(p_value)
                exploded = True
                break  # Exit the pressure loop once explosion occurs

        if exploded:
            break  # Move to next temperature after first explosion

# Plot the explosion curve
plt.figure()
plt.scatter(T_explosion, np.array(p_explosion) / ct.one_atm, label='Explosion Curve', color='red', marker='o')
plt.xlabel('Temperature [K]')
plt.ylabel('Pressure [atm]')
plt.title('Explosion Curve for Propane (C3H8)')
plt.legend()
plt.grid(True)
plt.show()
```
